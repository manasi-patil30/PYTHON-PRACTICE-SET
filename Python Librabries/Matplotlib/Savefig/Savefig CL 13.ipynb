# Savefig ()

```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4,5]
Y = [2,4,3,6,4]

plt.plot(X,Y, color = "r")
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [2,4,3,6,4]

plt.plot(X,Y, color = "r")

plt.savefig("Line")          # Save image using command

plt.show()
```

```python
X = [1,2,3,4,5]
Y = [2,4,3,6,4]

plt.plot(X,Y, color = "r")

plt.savefig("Line1.pdf", dpi = 1000, facecolor= "g")          # Change facecolor
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [2,4,3,6,4]

plt.plot(X,Y, color = "r")

plt.savefig("Line1.pdf", dpi = 1000, facecolor= "g", transparent = True)        # Save image using command

plt.show()
```

```python
X = [1,2,3,4,5]
Y = [2,4,3,6,4]

plt.plot(X,Y, color = "r")

plt.savefig("Line12.pdf", dpi = 1000, facecolor= "g", transparent = False, bbox_inches = "tight")        # Save image using command

plt.show()
```
