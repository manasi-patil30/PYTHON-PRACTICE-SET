```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\manas\Downloads\train.csv")
```

```python
df
```

```python
df1 =df.head(30)
df1
```

```python
df1.interpolate()          #  Using fill the empty value (not change data for string value)
```

```python
df1.interpolate(method = "linear")  #Change data linearly using method parameter
```

```python
df1.interpolate(method = "linear", axis = 0)  # Axis = 0 change the data column by column and axis = 1 change the data row by row
```

```python
# If we change the data row by row it is possible when all the rows are same data type("numerical Data Type")
# Even "data type" is not same as all rows we can't change the value
```

```python
df1.interpolate(limit = 2)       #  Give limit for change the data
```

```python
df1.interpolate(limit_direction = "forward", limit =2)   #  Give the direction for data we have to as forward/ backward
```

```python
df1.interpolate(limit_direction = "backward", limit =2) 
```

```python
df1.interpolate(limit_direction = "both", limit =2) 
```

```python
df1.interpolate(limit_area= "outside") 
```

```python
df1.interpolate(limit_area= "inside") 
```

```python
df1.interpolate(limit_direction = "both", limit =2,inplace = True) 
```
